```python
import seaborn as sns

sns.pairplot(bos) #Plots pairwise relationships in a dataset.
```

```python
sns.distplot(bos['Price']) #Flexibly plot a univariate distribution of observations.
```

```python
#Correlation Matrix
sns.set(style="white")
df_corr= bos[:]
corr = df_corr.dropna().corr() #Compute the correlation matrix
mask = np.zeros_like(corr, dtype=np.bool) #Generate a mask 
f, ax = plt.subplots(figsize=(30, 10)) #Set up the matplotlib figure
cmap = sns.diverging_palette(220, 10, as_cmap=True) #Generate a custom diverging colormap
sns.heatmap(corr, mask=mask, cmap=cmap, square=True, linewidths=.5, ax=ax) #Draw the heatmap with the mask and correct aspect ratio
#RAD and TAX are highly co-related.
#Price negatively corelated with LSTAT(Strong),PTRATIO(Strong),TAX(high), INDUS(High), CRIM(Highly) and 
#NOX highly corelated with RM.
#Also Price positively corelated with RM(High), ZN(High), CHAS(Medium), DIS(MEDIUM) & B(Medium)
```

```python
bos.isnull().values.any() #Check for any NAs in the dataframe.
```

```python
#Spliting the dataset into two: target value and predictor values. 
X = bos.drop('Price', axis = 1) #All features except Price ( predictor values )
Y = bos['Price'] #Boston housing price ( target value )
```

```python
import sklearn.model_selection
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X, Y, test_size = 0.25, random_state = 5)
print(X_train.shape) #Training data shape (predictor values) : 75%
print(X_test.shape) #Test data shape (predictor values) : 25%
print(Y_train.shape) #Training data shape (target values) : 75%
print(Y_test.shape) #Test data shape (target values) : 25%
```

```python
from sklearn.linear_model import LinearRegression #Imports LinearRegression
lm = LinearRegression() #Creating an instance of LinearRegression
lm.fit(X_train, Y_train) #Fitting the created instance of the LinearRegression
```

```python
#Printing intercept 
print(lm.intercept_)
```

```python
#Printing coefficients
print(lm.coef_)
```

```python
Y_pred = lm.predict(X_test) #Calculating the prediction values
```

```python

Y_pred.shape #Prediction shape from test data
```

```python
#Confidence in our Model
import statsmodels.formula.api as smf
lm = smf.ols(formula='Price ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT', data=bos).fit()
lm.conf_int() #Confidence Intervals of Model
```

```python
lm.summary() #Model Summary
```

```python
lm.pvalues #Prints the p-values (The probability that the coefficient is actually zero) for the model coefficients
```

```python
lm.rsquared #Prints the R-squared value for the model (Helps to check, "How Well Does the Model Fit the data?")
```

```python
#To visualize the differences between actual prices and predicted values, creating a scatter plot.
sns.set_style("whitegrid")
sns.set_context("poster")
plt.figure(figsize=(16,9))
plt.scatter(Y_test, Y_pred)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
plt.text(15,-20, "Comparison between the actual prices and predicted prices.", ha='left')
plt.show()
```

```python
sns.regplot(Y_test, Y_pred, data=bos, fit_reg=True) #Plot Y_test and Y_pred for Linear Regression Model.
```

```python
sns.regplot(x=lm.predict(X), y=bos['Price'], data=bos, fit_reg=True) #Plot predicted and actual Price values.
```

```python
#Calculating Mean Squared Error
mse = sklearn.metrics.mean_squared_error(Y_test, Y_pred) #Mean Squared Error: To check the level of error of a model
print(mse)
```

```python
#Calculating Root Mean Squared Error
rmse = mse ** 0.5 #Square root of mse (Mean Squared Error)
print(rmse)
```

```python
print("The Root Mean Square Error (RMSE) for the Model is "+ str(rmse) +" and the Results can be further improved using feature extraction and rebuilding, training the model.")
```
